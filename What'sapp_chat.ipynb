```python
import pandas as pd
import re
```

```python
with open("WhatsApp Chat with BS CS family group.txt", "r", encoding="utf-8") as file:
    data = file.read()
```

```python
print(data[:1000])
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?[APMapm]{2}\s-\s'

messages = re.split(pattern, data)[1:]
dates = re.findall(pattern, data)

print("Total Messages:", len(messages))
print("Total Dates:", len(dates))
```

```python
df = pd.DataFrame({
    "date": dates,
    "message": messages
})

df.head()
```

```python
df["date"] = df["date"].str.replace(" - ", "", regex=False)
```

```python
df["date"] = pd.to_datetime(
    df["date"],
    format="%m/%d/%y, %I:%M %p"
)
```

```python
df.head()
```

```python
users = []
messages = []

for message in df["message"]:
    entry = re.split(r"([^:]+):\s", message, maxsplit=1)

    if len(entry) >= 3:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("Group Notification")
        messages.append(entry[0])

df["user"] = users
df["message"] = messages
```

```python
df.info()
```

```python
# Make sure 'date' is datetime
df["date"] = pd.to_datetime(df["date"], format="mixed")

# Create new columns
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["month_name"] = df["date"].dt.month_name()
df["day"] = df["date"].dt.day
df["day_name"] = df["date"].dt.day_name()
df["hour"] = df["date"].dt.hour
df["minute"] = df["date"].dt.minute
df["date_only"] = df["date"].dt.date

# Create time period (e.g., 8-9 PM)
period = []
for hour in df["hour"]:
    start = hour
    end = (hour + 1) % 24
    period.append(f"{start:02d}-{end:02d}")

df["period"] = period
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urlextract import URLExtract
from collections import Counter
import emoji
from wordcloud import WordCloud

extract = URLExtract()
```

```python
def fetch_stats(selected_user, df):

    # If a specific user is selected
    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    # Total Messages
    num_messages = df.shape[0]

    # Total Words
    words = []
    for message in df['message']:
        words.extend(message.split())

    # Media Messages
    num_media_messages = df[df['message'] == '<Media omitted>\n'].shape[0]

    # Links Shared
    links = []
    for message in df['message']:
        links.extend(extract.find_urls(message))

    return num_messages, len(words), num_media_messages, len(links)
```

```python
num_messages, words, media, links = fetch_stats('Overall', df)

print("Total Messages :", num_messages)
print("Total Words :", words)
print("Media Shared :", media)
print("Links Shared :", links)
```

```python
def monthly_timeline(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    timeline = df.groupby(['year', 'month', 'month_name']).count()['message'].reset_index()

    time = []
    for i in range(timeline.shape[0]):
        time.append(f"{timeline['month_name'][i]}-{timeline['year'][i]}")

    timeline['time'] = time

    return timeline
```

```python
timeline = monthly_timeline('Overall', df)

timeline.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(timeline['time'], timeline['message'], marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Timeline")
plt.xlabel("Month")
plt.ylabel("Messages")
plt.grid(True)
plt.show()
```

```python
def daily_timeline(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    daily = df.groupby('only_date').count()['message'].reset_index()

    return daily
```

```python
# Convert date column to datetime
df["date"] = pd.to_datetime(df["date"], format="mixed")

# Extract all required features
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["month_name"] = df["date"].dt.month_name()
df["day"] = df["date"].dt.day
df["day_name"] = df["date"].dt.day_name()
df["hour"] = df["date"].dt.hour
df["minute"] = df["date"].dt.minute

# These names match the CampusX WhatsApp Chat Analyzer project
df["only_date"] = df["date"].dt.date

# Create period column (e.g., 13-14, 20-21)
period = []
for hour in df["hour"]:
    start = hour
    end = (hour + 1) % 24
    period.append(f"{start:02d}-{end:02d}")

df["period"] = period

# Verify
print(df.head())
print("\nColumns:")
print(df.columns.tolist())
```

```python
def activity_heatmap(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    heatmap = df.pivot_table(
        index='day_name',
        columns='period',
        values='message',
        aggfunc='count'
    ).fillna(0)

    return heatmap
```

```python
plt.figure(figsize=(18,6))

sns.heatmap(activity_heatmap('Overall',df))

plt.title("Weekly Activity Heatmap")
plt.show()
```

```python
stop_words = {
    'the','is','to','of','and','a','i','you','it','in',
    'for','on','this','that','with','my','me','your',
    'hai','ha','ki','ka','ke','ko','h','hy','hn',
    'ok','okay','yes','no'
}
```

```python
def create_wordcloud(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    temp = df[df['user'] != 'Group Notification']

    text = ""

    for message in temp['message']:
        for word in message.lower().split():
            if word not in stop_words:
                text += word + " "

    wc = WordCloud(
        width=900,
        height=500,
        background_color='white'
    ).generate(text)

    return wc
```

```python
wc = create_wordcloud('Overall',df)

plt.figure(figsize=(15,8))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
def most_common_words(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['user']==selected_user]

    temp = df[df['user']!='Group Notification']

    words=[]

    for message in temp['message']:
        for word in message.lower().split():
            if word not in stop_words:
                words.append(word)

    common = pd.DataFrame(Counter(words).most_common(20))

    return common
```

```python
common = most_common_words('Overall',df)

plt.figure(figsize=(10,8))
plt.barh(common[0],common[1])
plt.gca().invert_yaxis()
plt.title("Top 20 Most Common Words")
plt.show()
```

```python
def emoji_helper(selected_user, df):

    if selected_user != 'Overall':
        df=df[df['user']==selected_user]

    emojis=[]

    for message in df['message']:
        emojis.extend([c for c in message if c in emoji.EMOJI_DATA])

    emoji_df=pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis))))

    return emoji_df
```

```python
emoji_df = emoji_helper('Overall',df)

emoji_df.head(20)
```

```python
plt.figure(figsize=(10,5))
plt.bar(emoji_df[0][:10],emoji_df[1][:10])
plt.title("Top Emojis")
plt.show()
```

```python
def extract_links(selected_user,df):

    if selected_user!='Overall':
        df=df[df['user']==selected_user]

    links=[]

    for message in df['message']:
        links.extend(extract.find_urls(message))

    return pd.DataFrame(Counter(links).most_common(20),columns=['Link','Count'])
```

```python
def media_stats(selected_user,df):

    if selected_user!='Overall':
        df=df[df['user']==selected_user]

    media=df[df['message'].str.contains('<Media omitted>',na=False)]

    return media.shape[0]
```
